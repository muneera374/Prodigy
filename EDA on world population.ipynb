```python
# Distribution of 2023 population
sns.histplot(world["2023 population"],bins=10,kde=True)
plt.title("Distribution of 2023 population")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
#Continents according to highest population
world.groupby("continent")["2023 population"].mean().sort_values(ascending=False)
```

```python
# Top 10 countries with highest population in 2023
top_10_countries_2023=world.sort_values(by="2023 population",ascending=False).head(10)
```

```python
sns.barplot(x="2023 population",y="country",data=top_10_countries_2023)
plt.title("Top 10 countries by population in 2023")
plt.xlabel("population")
plt.ylabel("Country")
plt.show()
```

```python
world.columns
```

```python
sns.scatterplot(x="2023 population",y="density (km²)",data=top_10_countries_2023)
plt.title("Population and density scatterplot")
plt.xlabel("Population")
plt.ylabel("Density")
plt.show()
```

```python
world.groupby("continent")["area (km²)"].sum().sort_values(ascending=False)
```

```python
sns.violinplot(x="continent",y="density (km²)",data=world,palette="viridis")
plt.title("Population Density Distribution across Continents")
plt.xlabel("Continent")
plt.ylabel("Population Density")
plt.show()
```

```python
sns.boxplot(x="continent",y="density (km²)",data=world,palette="muted",showfliers=False)
plt.title("Population Density Distribution across Continents(Without Outliers)")
plt.xlabel("Continent")
plt.ylabel("Population Density")
plt.show()
```
