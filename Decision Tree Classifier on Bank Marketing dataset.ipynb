```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
bank_market_encoded=bank_market1.apply(lb.fit_transform)
bank_market_encoded
```

```python
bank_market_encoded["y"].value_counts()
```

Splitting the data into feature variables and class label

```python
from sklearn.model_selection import train_test_split
```

```python
fv=bank_market_encoded.drop("y",axis=1)
cv=bank_market_encoded["y"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(fv,cv,test_size=0.2,random_state=3,stratify=cv)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
def eval_model(y_test,y_pred):
    acc=accuracy_score(y_test,y_pred)
    print("Accuracy score:",acc)
    cm=confusion_matrix(y_test,y_pred)
    print("Confusion Matrix\n",cm)
    print("Classification Report\n",classification_report(y_test,y_pred))
    
def mscore(model):
    train_score=model.score(x_train,y_train)
    test_score=model.score(x_test,y_test)
    print("Training Score:",train_score)
    print("Testing Score:",test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="gini",max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
y_pred=dt.predict(x_test)
```

```python
y_pred
```

```python
eval_model(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
```
